# Sales Prediction 

In this analysis, we will predict sales using the Random Forest Regressor model. We will perform data loading, preprocessing, model training, evaluation, and visualization.

## Importing Necessary Libraries

We start by importing the required libraries for our analysis.




```python

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

# Data Loading
We load the advertising dataset, which contains information about advertising budgets and their corresponding sales.

```python
# Load the advertising dataset
df = pd.read_csv("Advertising.csv")
```

# Data Exploration
Basic Dataset Information
We display information about the dataset, including column names and data types.

```python
# Display basic dataset information
print("The DATASET")
print()
print(df.info())
```

# Top 10 Entries
We show the first 10 entries of the dataset.

```python
# Display the top 10 entries
print("Top 10 Entries:")
print(df.head(10))
```

# Data Splitting
We split the dataset into features (X) and the target (y) and further divide it into training and testing sets.

```python
# Split the dataset into features (X) and target (y)
X = df.drop(['Sales'], axis=1)
y = df['Sales']

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

# Model Training
We create and train a Random Forest Regressor model using the training data.

```python
# Create and train a Random Forest Regressor model
regressor = RandomForestRegressor(n_estimators=100, random_state=0)
regressor.fit(X_train, y_train)
```



# Sales Prediction
We use the trained model to make sales predictions.

```python
# Make sales predictions
y_prediction = regressor.predict(X_test)
```

# Model Evaluation
We evaluate the model's performance using the R-squared (R2) metric.

```python
# Evaluate the model
r_squared = r2_score(y_test, y_prediction)

# Display results
print(f"R-Squared: {r_squared}")
```



# Data Visualization
We visualize the distribution of sales and forecasted sales.

```python
# Visualize Sales distribution
plt.figure(figsize=(9, 6))
sns.distplot(df['Sales'])
plt.title("Sales Distribution")
plt.show()

# Visualize forecasted sales distribution
plt.figure(figsize=(8, 6))
sns.distplot(y_prediction)
plt.title("Forecasted Sales Distribution")
plt.show()

```

**This code uses the Random Forest Regressor model for sales prediction. It includes data loading, preprocessing, model training, evaluation, and visualization.**
